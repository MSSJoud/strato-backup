```python
import os
import glob
import re

# ---- Set your main paths ----
project_dir = "/mnt/data/gmtsar_test_1"
data_dir = os.path.join(project_dir, "data")
frames = ['F1', 'F2', 'F3']

# ---- Step 1: Clean broken or wildcard-based symlinks ----
for frame in frames:
    raw_dir = os.path.join(project_dir, frame, "raw")
    if not os.path.exists(raw_dir):
        continue
    for item in os.listdir(raw_dir):
        path = os.path.join(raw_dir, item)
        # Remove if broken link or a bad wildcard-pattern
        if os.path.islink(path) and (not os.path.exists(path) or '*' in item or '?' in item or re.search(r'\*|\?|\[\]', item)):
            print(f"ðŸ§¹ Removing broken or invalid link: {path}")
            os.unlink(path)

# ---- Step 2: Find valid EOF files ----
eof_files = glob.glob(os.path.join(data_dir, "S1[AC]_OPER_AUX_POEORB_*.EOF"))
print(f"\nðŸ”— Found {len(eof_files)} valid EOF files in {data_dir}")

# ---- Step 3: Relink .EOF files into each frame's raw directory ----
for frame in frames:
    frame_raw = os.path.join(project_dir, frame, "raw")
    os.makedirs(frame_raw, exist_ok=True)
    link_count = 0

    for eof_file in eof_files:
        target = os.path.join(frame_raw, os.path.basename(eof_file))
        if os.path.lexists(target):
            os.unlink(target)
        os.symlink(eof_file, target)
        link_count += 1

    print(f"âœ… {frame}/raw: Linked {link_count} EOF files")

# ---- Step 4: Verify symbolic links ----
print("\nðŸ” Verifying symbolic links in each frame's 'raw' directory:\n")
for frame in frames:
    frame_raw = os.path.join(project_dir, frame, "raw")
    broken_links = []
    total_links = 0

    for item in os.listdir(frame_raw):
        path = os.path.join(frame_raw, item)
        if os.path.islink(path):
            total_links += 1
            if not os.path.exists(path):
                broken_links.append(item)

    print(f"ðŸ“‚ {frame}/raw: {total_links} total links, {len(broken_links)} broken")
    if broken_links:
        print("âŒ Broken links:")
        for b in broken_links:
            print(f"   - {b}")
    else:
        print("âœ… All links valid.")
```

```python
from pathlib import Path
from datetime import datetime
import re

project_dir = Path("/mnt/data/gmtsar_test_1")
frames = {
    "F1": "iw1",
    "F2": "iw2",
    "F3": "iw3",
}

def extract_datetime_from_tiff(tiff_name):
    match = re.search(r"-(\d{8}t\d{6})-", tiff_name.lower())
    if not match:
        return None
    try:
        return datetime.strptime(match.group(1), "%Y%m%dt%H%M%S")
    except:
        return None

def orbit_covers(eof_name, dt):
    try:
        parts = eof_name.split("_V")[1].split("_")
        start = datetime.strptime(parts[0], "%Y%m%dT%H%M%S")
        end   = datetime.strptime(parts[1].replace(".EOF", ""), "%Y%m%dT%H%M%S")
        return start <= dt <= end
    except:
        return False

# Loop through each frame
for frame, iw in frames.items():
    raw_dir = project_dir / frame / "raw"
    tiffs = sorted(raw_dir.glob(f"*{iw}*vv*.tiff")) + sorted(raw_dir.glob(f"*{iw}*vh*.tiff"))
    eofs = sorted(raw_dir.glob("*.EOF"))
    data_in = raw_dir / "data.in"

    count = 0
    with open(data_in, "w") as f:
        for tiff in tiffs:
            dt = extract_datetime_from_tiff(tiff.name)
            if not dt:
                print(f"âš ï¸  Skipping {tiff.name} (no datetime found)")
                continue

            match = None
            for eof in eofs:
                if orbit_covers(eof.name, dt):
                    match = eof.name
                    break

            if match:
                f.write(f"{tiff.name}:{match}\n")
                count += 1
            else:
                print(f"âŒ No matching EOF for {tiff.name}")

    print(f"âœ… Created {data_in} with {count} entries.")
```
